```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial'
```

```python
date_list = [1, 2, 3, 4]
stock1 = [4, 8, 2, 6]
stock2 = [10, 12, 5, 3]

plt.plot(date_list, stock1)
plt.plot(date_list, stock2)
plt.show()
```

```python
plt.plot(date_list, stock1, 'ro--', label='StockId: Tencent')
plt.plot(date_list, stock2, 'b^--', label='StockId: Tesla')
plt.title('Stocking market')
plt.xlabel('time/day')
plt.ylabel('price')
plt.legend()
plt.show()
```
